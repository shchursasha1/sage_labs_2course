## Завдання 2. 
__Для кожної з визначених дій перевірити чи буде дія асоціативною. Якщо дія не асоціативна, то вивести трійку, для якої ця умова порушується.__

```python
def check_associativity(elements, operation):
    for a in elements:
        for b in elements:
            for c in elements:
                if operation((a, operation((b, c)))) != operation((operation((a, b)), c)):
                    return False, [a, b, c]
    return True, None

elements_set = [1, 2, 3]

valid_associative_operations = []
non_associative_triplets = []

for operation in all_actions(elements_set):
    is_associative, violating_triplet = check_associativity(elements_set, operation)

    if is_associative:
        valid_associative_operations.append(operation)
    elif violating_triplet not in non_associative_triplets:
        print("Found non-associative action with triplet:", violating_triplet)
        non_associative_triplets.append(violating_triplet)

print(f'Number of associative operations for a set of {len(elements_set)} elements:', len(valid_associative_operations))
```
